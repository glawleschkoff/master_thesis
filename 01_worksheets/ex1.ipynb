# Aufgabe 2

### Entropie

```python
H = lambda p : - sum([p(x) * np.log(p(x)) * 0.01 for x in np.arange(-100, 100, 0.01) if p(x) > 0])
```

##### Gleichverteilung

```python
p = lambda x : 1 if (x <= 1 and x >= 0) else 0
```

```python
H(p).item()
```

##### Gaussverteilung

```python
p = lambda sigma : lambda x : (1 / (2 * np.pi * sigma**2)**0.5) * np.exp(-(x - 0)**2 / (2 * sigma**2))
```

```python
H(p(4)).item()
```

### Kullback-Leibler Divergenz

```python
D = lambda q, p : sum([q(x) * np.log(q(x) / p(x)) * 0.01 for x in np.arange(-100, 100, 0.01) if (p(x) > 0 and q(x) > 0)])
```

```python
p = lambda mu : lambda sigma : lambda x : (1 / (2 * np.pi * sigma**2)**0.5) * np.exp(-(x - mu)**2 / (2 * sigma**2))
```

```python
D(p(0)(4.807), p(0)(1)).item()
```

```python
p_mm = lambda x : (p(0)(1)(x) + p(1)(1)(x)) / 2
```

```python
D(p(0.50)(1.118), p_mm).item()
```
