B[s<sup>y</sup><sub>t+1</sub>, s<sup>x</sup><sub>t+1</sub>, s<sup>y</sup><sub>t</sub>, s<sup>x</sup><sub>t</sub>] = p(s<sub>t+1</sub>|s<sub>t</sub>)

```python
B = np.zeros((4, 4, 4, 4))

B[0, 1, 0, 0] = 1

B[0, 0, 1, 0] = 0.5
B[1, 1, 1, 0] = 0.5

B[1, 0, 2, 0] = 0.5
B[2, 1, 2, 0] = 0.5

B[2, 0, 3, 0] = 0.5
B[3, 1, 3, 0] = 0.5

B[0, 2, 0, 1] = 1

B[0, 1, 1, 1] = 0.5
B[1, 2, 1, 1] = 0.5

B[1, 1, 2, 1] = 0.5
B[2, 2, 2, 1] = 0.5

B[2, 1, 3, 1] = 0.5
B[3, 2, 3, 1] = 0.5

B[0, 3, 0, 2] = 1

B[0, 2, 1, 2] = 0.5
B[1, 3, 1, 2] = 0.5

B[1, 2, 2, 2] = 0.5
B[2, 3, 2, 2] = 0.5

B[2, 2, 3, 2] = 0.5
B[3, 3, 3, 2] = 0.5

B[0, 3, 0, 3] = 1

B[0, 3, 1, 3] = 1

B[1, 3, 2, 3] = 1

B[2, 3, 3, 3] = 1
```

D[s<sup>y</sup>, s<sup>x</sup>] = p(s)

```python
D = np.zeros((4, 4))

D[2, 0] = 0.25
D[2, 1] = 0.25
D[3, 0] = 0.25
D[3, 1] = 0.25
```

```python
# helper functions

def safelog(x):
    return np.log(x + 1e-9)

def softmax(x):
    stabilized = x - np.max(x)
    exp_x = np.exp(stabilized)
    return exp_x / np.sum(exp_x)

def sample(prob):
    prob_flat = prob.ravel()
    sampled_index_prob_flat = np.random.choice(a=np.arange(len(prob_flat)), size=1, p=prob_flat)[0]
    (row_index, column_index) = np.unravel_index(sampled_index_prob_flat, prob.shape)
    return (row_index, column_index)
```

```python
def generative_process(A, B, D, T):

    observables = np.zeros((T, 2), dtype=int)
    hidden = np.zeros((T, 2), dtype=int)
    hidden[0] = sample(D)
    for i in range(T):
        observables[i, :] = sample(A[:, :, hidden[i, 0], hidden[i, 1]])
        if i < T - 1:
            hidden[i + 1, :] = sample(B[:, :, hidden[i, 0], hidden[i, 1]])

    return hidden, observables
```

```python
def print_generative_process(hidden, observables):
    plt.figure(figsize=(4, 4))
    plt.title("Generativer Prozess", fontsize=12)
    x_hidden = hidden[:, 1]
    y_hidden = hidden[:, 0]
    plt.plot(x_hidden, y_hidden, marker='o', linestyle='-', color='b', label="hidden")
    x_observables = observables[:, 1]
    y_observables = observables[:, 0]
    plt.plot(x_observables, y_observables, marker='.', linestyle='-', color='c', label="observable")
    for i in range(len(x_hidden)):
        plt.text(x_hidden[i] + 0.08, y_hidden[i] - 0.02, str(i + 1), color='blue', fontsize=10)
    for i in range(len(x_observables)):
        plt.text(x_observables[i] + 0.08, y_observables[i] + 0.14, str(i + 1), color='darkcyan', fontsize=10)

    # Achsen und Legende
    plt.xticks([0, 1, 2, 3])
    plt.yticks([0, 1, 2, 3])
    #plt.gca().invert_yaxis()
    plt.legend()
    plt.grid(True, alpha=0.2)
    plt.show()
```

```python
def infer_meanfield(observables, A, B, D, T, REP):
    N_observables = len(observables)
    q_hidden = np.ones((T, 4, 4)) / 16
    q_obs = np.ones((T - N_observables, 4, 4)) / 16

    for rep in range(REP):
        for t in range(T):
            if T == 1:
                if N_observables > 0:
                    o = (observables[t, 0], observables[t, 1])
                    q_hidden[t] = softmax(safelog(A[o[0], o[1]]) + safelog(D))
                else:
                    q_hidden[t] = softmax(np.einsum('ij,ijmn->mn', q_obs[0], safelog(A)) + safelog(D))
                    q_obs[t] = softmax(np.einsum('ij,mnij->mn', q_hidden[0], safelog(A)))
            else:
                if t == 0:
                    if N_observables > t:
                        o = (observables[t, 0], observables[t, 1])
                        q_hidden[t] = softmax(np.einsum('ij,ijmn->mn', q_hidden[t + 1], safelog(B)) + safelog(A[o[0], o[1]]) + safelog(D))
                    else:
                        q_hidden[t] = softmax(np.einsum('ij,ijmn->mn', q_hidden[t + 1], safelog(B)) + np.einsum('ij,ijmn->mn', q_obs[0], safelog(A)) + safelog(D))
                        q_obs[t] = softmax(np.einsum('ij,mnij->mn', q_hidden[0], safelog(A)))
                elif t == T - 1:
                    if N_observables > t:
                        o = (observables[t, 0], observables[t, 1])
                        q_hidden[t] = softmax(np.einsum('ij,mnij->mn', q_hidden[t - 1], safelog(B)) + safelog(A[o[0], o[1]]))
                    else:
                        q_hidden[t] = softmax(np.einsum('ij,mnij->mn', q_hidden[t - 1], safelog(B)) + np.einsum('ij,ijmn->mn', q_obs[t - N_observables - 1], safelog(A)))
                        q_obs[t - N_observables - 1] = softmax(np.einsum('ij,mnij->mn', q_hidden[t], safelog(A)))
                else:
                    if N_observables > t:
                        o = (observables[t, 0], observables[t, 1])
                        q_hidden[t] = softmax(np.einsum('ij,ijmn->mn', q_hidden[t + 1], safelog(B)) + safelog(A[o[0], o[1]]) + np.einsum('ij,mnij->mn', q_hidden[t - 1], safelog(B)))
                    else:
                        q_hidden[t] = softmax(np.einsum('ij,ijmn->mn', q_hidden[t + 1], safelog(B)) + np.einsum('ij,ijmn->mn', q_obs[t - N_observables - 1], safelog(A)) + np.einsum('ij,mnij->mn', q_hidden[t - 1], safelog(B)))
                        q_obs[t - N_observables - 1] = softmax(np.einsum('ij,mnij->mn', q_hidden[t], safelog(A)))
                    
    return q_hidden, q_obs
```

```python
def visualize(generated_hidden, generated_obs, q_hidden, q_obs):
    T = len(q_hidden)
    fig, ax = plt.subplots(1, T, figsize=(T * 2.6 + 3, 3.5), sharey=True)
    if T == 1:
        ax = [ax]
    for t in range(T):
        ax[t].xaxis.tick_top()
        ax[t].set_xlabel(f"t = {t + 1}")
        ax[t].imshow(np.zeros((4, 4)), cmap='Greys', vmin=0, vmax=1)
        ax[t].set_xticks(np.arange(-.5, 4, 1), minor=True)
        ax[t].set_yticks(np.arange(-.5, 4, 1), minor=True)
        ax[t].grid(which='minor', color='0.1', linestyle='-', linewidth=1) # Helles Gitter
        ax[t].set_aspect('equal')
    for t in range(len(generated_hidden)):
        ax[t].scatter(generated_hidden[t, 1], generated_hidden[t, 0],
                      s=150,
                      marker='x',
                      color='#FFBF00',
                      linewidth=2.5,
                      zorder=6)
        ax[t].scatter(generated_hidden[t, 1], generated_hidden[t, 0],
                      s=180,
                      marker='x',
                      color='white',
                      linewidth=4,
                      zorder=4)
        ax[t].scatter(generated_obs[t, 1], generated_obs[t, 0],
                      s=600,
                      marker='o',
                      color='#FFBF00',
                      zorder=3)
    for t in range(T):
        q = q_hidden[t]
        for y in range(4):
            for x in range(4):
                if q[y, x] >= 0.01:
                    ax[t].scatter(x, y,
                                  s=200,
                                  marker='x',
                                  color=f"{1 - q[y, x]:.2f}",
                                  linewidth=6,
                                  zorder=5)
                    
    for t in range(T - len(q_obs), T):
        q = q_obs[t - (T - len(q_obs))]
        for y in range(4):
            for x in range(4):
                if q[y, x] >= 0.01:
                    ax[t].scatter(x, y,
                                  s=700,
                                  marker='o',
                                  color=f"{1 - q[y, x]:.2f}",
                                  linewidth=3.5,
                                  zorder=1)
                    ax[t].scatter(x, y,
                                  s=500,
                                  marker='o',
                                  color='white',
                                  zorder=2)

    legend_elements = [
        Line2D([0], [0], marker='x', color='#FFBF00', linestyle='None',
               markersize=10, markeredgewidth=2.5),
        Line2D([0], [0], marker='o', color='#FFBF00', linestyle='None',
               markersize=12, markerfacecolor='#FFBF00', markeredgecolor='#FFBF00'),
        Line2D([0], [0], color='white'),
        Line2D([0], [0], marker='x', color='white', linestyle='None',
               markersize=10, markeredgecolor='0.5', markeredgewidth=4),
        Line2D([0], [0], marker='o', color='none', linestyle='None',
               markersize=12, markerfacecolor='none', markeredgecolor='0.7', markeredgewidth=3)
    ]
    legend_labels = [
        'Wahrer Zustand (generiert)',
        'Wahre Beobachtung (generiert)',
        '',
        'Inferierter Zustand (q_hidden)',
        'Inferierte Beobachtung (q_obs)'
    ]
    fig.legend(handles=legend_elements, 
               labels=legend_labels,
               loc='center left',
               bbox_to_anchor=(0.9, 0.65),
               fontsize='medium',
               frameon=False)

    cax = fig.add_axes([0.93, 0.44, 0.12, 0.04])
    cmap = plt.get_cmap('Greys') 
    norm = mcolors.Normalize(vmin=0, vmax=1)
    sm = cm.ScalarMappable(norm=norm, cmap=cmap)
    cbar = fig.colorbar(sm, cax=cax, orientation='horizontal')
    cbar.set_label('Wahrscheinlichkeit (p)', fontsize='small')
    cbar.set_ticks([0, 1])
    cbar.set_ticklabels(['0 (Weiß)', '1 (Schwarz)'], fontsize='small')
    cbar.outline.set_visible(False)
```

```python
generated_hidden, generated_obs = generative_process(A, B, D, T=2)
q_hidden, q_obs = infer_meanfield(generated_obs, A, B, D, T=5, REP=1000)
visualize(generated_hidden, generated_obs, q_hidden, q_obs)
```
