```python
import numpy as np
```

#### $p(s_1,s_2,s_3,o_1,o_2,o_3)=p(s_1)p(s_2|s_1)p(s_3|s_2)p(o_1|s_1)p(o_2|s_2)p(o_3|s_3)$ **(Model)**

```python
p_D = np.zeros(2)
p_D[0] = 0.4
p_D[1] = 0.6

p_B = np.zeros((2, 2))
p_B[0, 0] = 0.2
p_B[0, 1] = 0.7
p_B[1, 0] = 0.8
p_B[1, 1] = 0.3

p_A = np.zeros((2, 2))
p_A[0, 0] = 0.1
p_A[0, 1] = 0.5
p_A[1, 0] = 0.9
p_A[1, 1] = 0.5
```

#### Analytical

```python
p_s1_s2_s3_o1_o2_o3 = np.einsum('a,ba,cb,da,eb,fc->abcdef', p_D, p_B, p_B, p_A, p_A, p_A)

p_s1 = np.einsum('abcdef->a', p_s1_s2_s3_o1_o2_o3)
p_s2 = np.einsum('abcdef->b', p_s1_s2_s3_o1_o2_o3)
p_s3 = np.einsum('abcdef->c', p_s1_s2_s3_o1_o2_o3)
p_o1 = np.einsum('abcdef->d', p_s1_s2_s3_o1_o2_o3)
p_o2 = np.einsum('abcdef->e', p_s1_s2_s3_o1_o2_o3)
p_o3 = np.einsum('abcdef->f', p_s1_s2_s3_o1_o2_o3)

print('p(s_1) = ' + str(p_s1))
print('p(s_2) = ' + str(p_s2))
print('p(s_2) = ' + str(p_s3))
print('p(o_1) = ' + str(p_o1))
print('p(o_2) = ' + str(p_o2))
print('p(o_3) = ' + str(p_o3))
```

#### Sum Product

```python
r_mu_s1 = p_D
u_mu_sss1 = np.einsum('os->s', p_A)
r_mu_ss1 = np.einsum('s,s->s', r_mu_s1, u_mu_sss1)
r_mu_s2 = np.einsum('ab,b->a', p_B, r_mu_ss1)
u_mu_sss2 = np.einsum('os->s', p_A)
r_mu_ss2 = np.einsum('s,s->s', r_mu_s2, u_mu_sss2)
r_mu_s3 = np.einsum('ab,b->a', p_B, r_mu_ss2)

u_mu_sss3 = np.einsum('os->s', p_A)
l_mu_s3 = u_mu_sss3
l_mu_ss2 = np.einsum('ab,a->b', p_B, l_mu_s3)
l_mu_s2 = np.einsum('s,s->s', l_mu_ss2, u_mu_sss2)
l_mu_ss1 = np.einsum('ab,a->b', p_B, l_mu_s2)
l_mu_s1 = np.einsum('s,s->s', l_mu_ss1, u_mu_sss1)

d_mu_sss1 = np.einsum('s,s->s', r_mu_s1, l_mu_ss1)
d_mu_o1 = np.einsum('os,s->o', p_A, d_mu_sss1)
d_mu_sss2 = np.einsum('s,s->s', r_mu_s2, l_mu_ss2)
d_mu_o2 = np.einsum('os,s->o', p_A, d_mu_sss2)
d_mu_sss3 = r_mu_s3
d_mu_o3 = np.einsum('os,s->o', p_A, d_mu_sss3)

q_s1 = (r_mu_s1 * l_mu_s1) / np.sum(r_mu_s1 * l_mu_s1)
q_s2 = (r_mu_s2 * l_mu_s2) / np.sum(r_mu_s2 * l_mu_s2)
q_s3 = (r_mu_s3 * l_mu_s3) / np.sum(r_mu_s3 * l_mu_s3)
q_o1 = (d_mu_o1) / np.sum(d_mu_o1)
q_o2 = (d_mu_o2) / np.sum(d_mu_o2)
q_o3 = (d_mu_o3) / np.sum(d_mu_o3)

print('q(s_1) = ' + str(q_s1))
print('q(s_2) = ' + str(q_s2))
print('q(s_3) = ' + str(q_s3))
print('q(o_1) = ' + str(q_o1))
print('q(o_2) = ' + str(q_o2))
print('q(o_3) = ' + str(q_o3))
```
