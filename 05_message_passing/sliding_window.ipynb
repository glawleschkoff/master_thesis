```python
import numpy as np
```

```python
p_D = np.zeros(2)
p_D[0] = 0.4
p_D[1] = 0.6

p_B = np.zeros((2, 2))
p_B[0, 0] = 0.2
p_B[0, 1] = 0.7
p_B[1, 0] = 0.8
p_B[1, 1] = 0.3

p_A = np.zeros((2, 2))
p_A[0, 0] = 0.1
p_A[0, 1] = 0.5
p_A[1, 0] = 0.9
p_A[1, 1] = 0.5
```

```python
p_s0_s1_s2_o1 = np.einsum('i,ji,kj,lj->ijkl', p_D, p_B, p_B, p_A)
p_s0 = np.einsum('ijkl->i', p_s0_s1_s2_o1)
p_s1 = np.einsum('ijkl->j', p_s0_s1_s2_o1)
p_s2 = np.einsum('ijkl->k', p_s0_s1_s2_o1)
p_o1 = np.einsum('ijkl->l', p_s0_s1_s2_o1)

print('p(s_0): ' + str(p_s0))
print('p(s_1): ' + str(p_s1))
print('p(s_2): ' + str(p_s2))
print('p(o_1): ' + str(p_o1))
```

```python
o = 0

u_mu_s1 = p_A[o]
#u_mu_s1 = np.einsum('ij->j', p_A)
r_mu_s0 = p_D
r_mu_s1 = np.einsum('ij,j->i', p_B, r_mu_s0)
r_mu_s2 = np.einsum('ij,j,j->i', p_B, r_mu_s1, u_mu_s1)
l_mu_s1 = np.einsum('ij->j', p_B)
l_mu_s0 = np.einsum('ij,i,i->j', p_B, l_mu_s1, u_mu_s1)

q_s0 = (r_mu_s0 * l_mu_s0) / np.sum(r_mu_s0 * l_mu_s0)
q_s1 = (r_mu_s1 * u_mu_s1 * l_mu_s1) / np.sum(r_mu_s1 * u_mu_s1 * l_mu_s1)
q_s2 = (r_mu_s2) / np.sum(r_mu_s2)

print('q(s_0): ' + str(q_s0))
print('q(s_1): ' + str(q_s1))
print('q(s_2): ' + str(q_s2))
```
