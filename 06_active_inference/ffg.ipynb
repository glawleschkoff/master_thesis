```python
import plotly.graph_objects as go

fig = go.Figure()

# Das Rechteck zeichnen
fig.add_shape(
    type="rect",
    x0=0, y0=0,  # Koordinaten der unteren linken Ecke
    x1=1, y1=1,  # Koordinaten der oberen rechten Ecke
    line=dict(color="black", width=1), # Schwarzer Rand
)
fig.add_shape(
    type="rect",
    x0=2, y0=0,  # Koordinaten der unteren linken Ecke
    x1=3, y1=1,  # Koordinaten der oberen rechten Ecke
    line=dict(color="black", width=1), # Schwarzer Rand
)
fig.add_shape(
    type="line",
    x0=1, y0=0.5,
    x1=2, y1=0.5,
    line=dict(color="black", width=1), # Schwarzer Rand
)
fig.add_annotation(
    x=0.5,           # Die exakte Mitte zwischen x0=0 und x1=1
    y=0.5,           # Die exakte Mitte zwischen y0=0 und y1=1
    text=r"$p(s_0)$", # Dein LaTeX-Code!
    showarrow=False, # Ganz wichtig, sonst zeichnet Plotly einen Pfeil zum Text
    font=dict(size=18, color="black")
)

# Das Fenster einrichten, damit das Rechteck in die Ansicht passt
fig.update_layout(
    xaxis=dict(range=[-1, 4], visible=False), # visible=False versteckt die Achsen
    yaxis=dict(range=[-1, 2], visible=False),
    width=600, height=400,
    plot_bgcolor='white'
)

# <-- HIER IST DIE WICHTIGE ZERSETZLICHE ZEILE -->
# Zwingt Achsenverhältnis 1:1, damit 1 Einheit X = 1 Einheit Y (in Pixeln)
fig.update_yaxes(scaleanchor="x", scaleratio=1) 

fig.show(renderer="svg")
```
