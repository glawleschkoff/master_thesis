```python
import numpy as np
import matplotlib.pyplot as plt

delta_t = 0.01
T = 5
steps = int(T / delta_t)

phi = np.zeros(steps)
epsilon_p = np.zeros(steps)
epsilon_u = np.zeros(steps)
phi[0] = 3
v_p = 3
u = 2
sigma_p = 1
sigma_u = 1
```

```python
for t in range(steps - 1):
    g_prime = 2 * phi[t]
    g = phi[t] ** 2
    phi[t + 1] = phi[t] + delta_t * (epsilon_u[t] * g_prime - epsilon_p[t])
    epsilon_p[t + 1] = epsilon_p[t] + delta_t * (phi[t] - v_p - sigma_p * epsilon_p[t])
    epsilon_u[t + 1] = epsilon_u[t] + delta_t * (u - g - sigma_u * epsilon_u[t])
```

```python
x = np.arange(steps) / 100

fig, ax = plt.subplots()
ax.plot(x, phi, label='$\phi$')
ax.plot(x, epsilon_p, label='$\epsilon_p$')
ax.plot(x, epsilon_u, label='$\epsilon_u$')
ax.legend()
ax.set_xlabel('Time')
ax.set_ylabel('Activity')
plt.show()
```
